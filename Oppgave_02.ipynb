La variablene salgspris per enhet, antall solgte enheter og kostnader henholdsvis få verdiene 125, 750 og 100000. Undersøk om overskuddet er positivt ved at Python returnerer enten True eller False.

```python
# Gir verdi til variablene
salgspris = 125
antall = 750
kostnader = 100000
```

```python
# Beregner overskudd, og undersøker om det er positivt
overskudd = salgspris * antall - kostnader
overskudd > 0
```
