# Oppgave 3
Legg inn varene Brød, Kaffe og Melk som streng-variabler, og la variablene for de tilhørende prisene 214.51, 1283.25 og 44.70 være flyttall. Bruk deretter print og f-streng til å lage den flotte tabellen.

```python
# Gir verdi til variablene
vare_1, vare_2, vare_3 = 'Brød', 'Kaffe', 'Melk'
pris_1, pris_2, pris_3 = 214.51, 1283.25, 44.70
```

```python
# Lager tabellen
print(f'{'Vare':10}{'Pris':^8}')
print('-'*18)
print(f'{vare_1:10}{'kr'}{pris_1:>6.0f}')
print(f'{vare_2:10}{'kr'}{pris_2:>6.0f}')
print(f'{vare_3:10}{'kr'}{pris_3:>6.0f}')
```
