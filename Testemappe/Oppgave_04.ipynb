# Oppgave 4
Lag en (handle)liste med Shampoo, Balsam, Fløte, Lettmelk, Epler og Løk i rekkefølge. Gjør så følgende: Erstatt Lettmelk med Helmelk, fjern Shampoo og Balsam, og legg til Brokkoli mellom Epler og Løk.

```python
# Lager handlelisten
handleliste = ['Shampoo', 'Balsam', 'Fløte', 'Lettmelk', 'Epler', 'Løk']
handleliste
```

```python
# Erstatter Lettmelk med Helmelk
handleliste[3] = 'Helmelk'
handleliste
```

```python
# Fjerner Shampoo og Balsam
del handleliste[0:2]
handleliste
```

```python
# Legger til Brokkoli mellom Epler og Løk
handleliste.insert(3, 'Brokkoli')
handleliste
```
