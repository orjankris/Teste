# Oppgave 6
La variabelen pris være parkeringsavgift per time, og la variablene start og stopp på formen 'hh:mm' være start og avslutning på parkeringen. Hva koster det å parkere fra kl 11:27 til kl 15:03 når timeprisen er 32 kr?

```python
# Inndata
pris = 32
start, stopp = '11:27', '15:03'

# Konverterer til heltall, og gjør om til timer
fra = int(start[0:2]) + int(start[3:])/60
til = int(stopp[0:2]) + int(stopp[3:])/60

# Beregner hva det koster å parkere
kostnad = (til-fra) * pris
print(f'Det koster {kostnad:.2f} kr å parkere fra {start} til {stopp}.')
```
